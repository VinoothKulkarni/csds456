```python
twitterusers = pd.read_csv("gdrive/My Drive/Data_Privacy_Project/twitterusers.csv")
twitterusers_en = twitterusers[twitterusers['language']=='en']
twitterusers_en.index
```

```python
gender_similarity = pd.read_csv("gdrive/My Drive/Data_Privacy_Project/gender_similarity.csv",index_col=0)
gender_similarity = gender_similarity.iloc[twitterusers_en.index]
gender_similarity = gender_similarity.iloc[:, twitterusers_en.index]
gender_similarity = gender_similarity.reset_index(drop=True)
gender_similarity
```

```python
name_similarity = pd.read_csv("gdrive/My Drive/Data_Privacy_Project/name_similarity.csv",index_col=0)
name_similarity = name_similarity.iloc[twitterusers_en.index]
name_similarity = name_similarity.iloc[:, twitterusers_en.index]
name_similarity = name_similarity.reset_index(drop=True)
name_similarity
```

Random Sampling

```python
import random
list_select = random.sample(range(2311), 1500)
```

```python
name_similarity.iloc[1][2]
```

```python
import pandas as pd
  
# create an Empty DataFrame object With
# column names and indices 
df = pd.DataFrame(columns = ['Name_Similarity', 'Gender_Similarity', 'Matched'])

for i in list_select:
  if i > 2305:
    j = i-5
  else:
    j = i+5
  df = df.append({'Name_Similarity' : name_similarity.iloc[i][i], 'Gender_Similarity': gender_similarity.iloc[i][i], 'Matched' : 1}, ignore_index = True)
  df = df.append({'Name_Similarity' : name_similarity.iloc[i][j], 'Gender_Similarity': gender_similarity.iloc[i][j], 'Matched' : 0}, ignore_index = True)
```

```python
df.to_csv("Training_data.csv",index=False)
```
